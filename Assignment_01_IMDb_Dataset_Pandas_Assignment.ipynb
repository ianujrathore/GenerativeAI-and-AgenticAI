### Question 10
Create a report showing Total Movies, Average Budget, Average Revenue, Average Profit, Maximum Profit, Minimum Profit.

```python
# Write your solution here
Total_Movies = df['original_title'].count()
Budget = df['budget'].mean()
Revenue = df['revenue'].mean()
Profit_Average = df['profit'].mean()
Max_Profit = df['profit'].max()
Min_Profit = df['profit'].min()

print('Total Movies:',Total_Movies)
print('Average Budget:',Budget)
print('Average Revenue:',Revenue)
print('Average Profit:',Profit_Average)
print('Maximum Profit:',Max_Profit)
print('Minimum Profit:',Min_Profit)
```
