#Q2.This data has information about movies made in different languages. Which language has the highest average ROI (return on investment)?

```python
df.describe()
```

```python
df['budget'] = df['budget'].apply(lambda x: df['budget'].median() if x < 1000 else x)
df['roi'] = (df['revenue'] - df['budget'])/df['budget']
df.groupby('original_language')['roi'].mean().sort_values(ascending=False).head(1)
```

#Q5.Which actor has acted in the most number of movies? Deep dive into the movies, genres and profits corresponding to this actor.

```python
df.info()
```

```python
from collections import Counter
counter=Counter()
for it in df['cast'].dropna():
  for name in eval(it):
    counter[name['name']]+=1
counter.most_common(1)
best_actor=counter.most_common(1)[0][0]
print("Actor with most number of movies : ",best_actor)
print("Number of movies : ",counter.most_common(1)[0][1])
```

```python
#Details of all movies actor is in
best_actor_movies=df[df['cast'].apply(lambda x:pd.notna(x) and best_actor in x)]
best_actor_movies
```

```python
#genres
best_actor_movies=best_actor_movies.copy()
best_actor_movies['genre_names']=best_actor_movies['genres'].apply(lambda x: " ".join(it['name'] for it in eval(x)))
best_actor_movies[['original_title','genre_names','release_date','budget','revenue','profit']]
```

```python
#most profitable movie
best_actor_movies[['original_title','profit']].sort_values('profit',ascending=False).head(1)
```

```python
print("Least profitable movie :")
best_actor_movies[['original_title','profit']].sort_values('profit',ascending=True).head(1)
```

```python
#most repeated genre

genres=[]
for genre in best_actor_movies['genre_names']:
  genres.append(genre.split())
counter=Counter()
for g in genres:
  for it in g:
    counter[it]+=1
counter.most_common(1)
print(best_actor)
print("Most repeated genre : ",counter.most_common(1)[0][0])
print("Number of movies : ",counter.most_common(1)[0][1])
```

```python
best_actor_movies['release_date'].head()
```

```python
#movies across years
best_actor_movies=best_actor_movies.copy()
best_actor_movies['release_year']=pd.to_datetime(best_actor_movies['release_date'],format='%m/%d/%y',errors='coerce').dt.year
best_actor_movies.groupby(best_actor_movies['release_year']).size()
```

```python
best_actor_movies.groupby(best_actor_movies['release_year']).size().plot(
    kind='bar',
    color = 'blue',
    figsize=(16,6),
    title=f'{best_actor} Movies per year',
    xlabel='release_years',
    ylabel='number_of_movies'
)
```
