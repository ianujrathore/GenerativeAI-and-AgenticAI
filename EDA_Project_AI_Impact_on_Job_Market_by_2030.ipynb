## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
# Chart - 1 visualization code
plt.figure(figsize=(8,3))
sns.countplot(x='Industry', data=df, order=df['Industry'].value_counts().index)
plt.title('Distribution of Employees Across Industries')
plt.xlabel('Industry')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()
````

##### 1. Why did you pick the specific chart?

To understand how employee records are distributed across different industries and identify the industries with the highest representation.

##### 2. What is/are the insight(s) found from the chart?

The chart shows the number of employees in each industry.
It helps identify the most and least represented industries in the dataset.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes. Understanding industry distribution helps identify sectors where AI adoption and workforce planning may have the greatest impact.
No direct negative business impact is observed from this distribution analysis.

#### Chart - 2

````python
# Chart - 2 visualization code
plt.figure(figsize=(8,3))
sns.countplot(x='Education_Level', data=df, order=df['Education_Level'].value_counts().index)
plt.title('Distribution of Education Levels')
plt.xlabel('Education Level')
plt.ylabel('Number of Employees')
plt.show()
````

##### 1. Why did you pick the specific chart?

To understand the distribution of employees across different education levels and identify the most common qualification in the dataset.

##### 2. What is/are the insight(s) found from the chart?

The chart shows the frequency of each education level among employees.
It helps identify whether the workforce is dominated by a particular qualification or is evenly distributed.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes. Understanding employees' education levels can help organizations design targeted hiring, training, and upskilling strategies.
No direct negative business impact is observed, but an imbalance in education levels may indicate the need for workforce diversification.

#### Chart - 3

````python
# Chart - 3 visualization
plt.figure(figsize=(8,3))
sns.histplot(df['AI_Replacement_Risk'], bins=20, kde=True)

plt.title('Distribution of AI Replacement Risk')
plt.xlabel('AI Replacement Risk')
plt.ylabel('Frequency')
plt.show()
````

##### 1. Why did you pick the specific chart?

To understand how AI replacement risk is distributed across all employees and identify whether most jobs have low, moderate, or high risk.

##### 2. What is/are the insight(s) found from the chart?

The chart shows the distribution of AI replacement risk across the workforce.
It helps identify whether AI risk is concentrated around specific values or spread evenly.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes. Organizations can identify the overall level of AI risk and prioritize workforce planning and upskilling initiatives.
Jobs with consistently high AI replacement risk may require reskilling programs to reduce future employment challenges.

#### Chart - 4

````python
# Chart - 4 visualization code
plt.figure(figsize=(8,3))
sns.boxplot(x=df['Average_Salary_USD'], color='mediumpurple')

plt.title('Distribution of Average Salary (USD)')
plt.xlabel('Average Salary (USD)')
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A boxplot is useful for understanding the distribution of salaries and identifying the presence of outliers.Answer Here.

##### 2. What is/are the insight(s) found from the chart?

The chart shows the spread, median, and variability of employee salaries.
It also highlights any unusually high or low salary values in the dataset.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes. Understanding salary distribution helps organizations benchmark compensation and identify potential pay disparities.
Significant salary outliers may indicate inconsistencies in compensation policies and warrant further investigation.

#### Chart - 5

````python
# Chart - 5 visualization code
plt.figure(figsize=(8,3))
sns.violinplot(x=df['Job_Satisfaction'], color='lightseagreen')
plt.title('Distribution of Job Satisfaction')
plt.xlabel('Job Satisfaction')
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A violin plot shows both the distribution and density of job satisfaction scores, providing more information than a simple boxplot.

##### 2. What is/are the insight(s) found from the chart?

The chart reveals how job satisfaction scores are distributed and where employee responses are most concentrated.
It also highlights the spread and variation in satisfaction levels.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes. Understanding employee satisfaction helps organizations improve workplace policies, employee engagement, and retention.
Lower satisfaction levels may indicate areas where organizations need to focus on improving the work environment and employee well-being.
