# **Project Summary -**

The Indian Premier League (IPL) is one of the most popular and competitive Twenty20 (T20) cricket leagues in the world. Since its inception in 2008, the tournament has featured some of the best domestic and international players, making it an excellent source of sports data for analysis. Every match generates valuable information related to teams, venues, toss decisions, match results, players, and other match-specific details. Analyzing this data helps uncover meaningful patterns and provides a deeper understanding of the game beyond the final score.

This project focuses on performing Exploratory Data Analysis (EDA) on the IPL Matches dataset, which contains match-level information from the 2008 to 2016 IPL seasons. The primary objective of the project is to explore the dataset, understand its structure, clean and preprocess the data, and identify important trends and relationships using statistical summaries and data visualization techniques. Rather than making predictions, the project aims to understand the characteristics of the data and extract useful insights that can support informed decision-making and further analysis.

The analysis begins with data inspection to understand the dataset's features, data types, missing values, and overall quality. Appropriate preprocessing steps, such as handling missing values, converting data types, and removing unnecessary columns, are carried out to prepare the data for analysis. Once the data is cleaned, various exploratory techniques are applied to examine different aspects of IPL matches. These include season-wise match distribution, team performances, toss outcomes, match winners, winning margins, venue analysis, city-wise match distribution, and Player of the Match awards. Visualizations such as bar charts, count plots, histograms, box plots, and heatmaps are used to present the findings in a clear and meaningful manner.

The project also aims to answer practical questions, such as whether winning the toss influences the match result, which teams have been the most successful over the years, which venues have hosted the highest number of matches, and how victory margins vary across matches. By interpreting these observations, the project highlights important trends and patterns present in the IPL dataset.

Overall, this project demonstrates the practical application of Exploratory Data Analysis in understanding real-world sports data. It emphasizes the importance of data cleaning, visualization, and statistical exploration as essential steps in the data analysis process. The insights generated through this study provide a comprehensive overview of IPL matches and serve as a foundation for future work, such as predictive modeling, performance evaluation, and sports analytics.

## ***2. Understanding Your Variables***

````python
# Dataset Columns
df.columns
````

````python
# Dataset Describe
df.describe()
````

### Variables Description

* **id:** Unique identifier assigned to each IPL match.
* **season:** IPL season in which the match was played.
* **city:** City where the match took place.
* **date:** Date on which the match was played.
* **team1:** First team participating in the match.
* **team2:** Second team participating in the match.
* **toss_winner:** Team that won the toss.
* **toss_decision:** Decision taken by the toss-winning team (bat or field).
* **result:** Final status of the match (normal, tie, or no result).
* **dl_applied:** Indicates whether the Duckworth-Lewis method was applied (1 = Yes, 0 = No).
* **winner:** Team that won the match.
* **win_by_runs:** Margin of victory in runs when the team batting first won.
* **win_by_wickets:** Margin of victory in wickets when the team batting second won.
* **player_of_match:** Player awarded for the best performance in the match.
* **venue:** Stadium where the match was played.
* **umpire1:** First on-field umpire.
* **umpire2:** Second on-field umpire.
* **umpire3:** Third umpire column, which contains only missing values in this dataset.
