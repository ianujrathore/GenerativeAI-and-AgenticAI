```python
dl_model = Sequential([
    Dense(16, activation='relu', input_shape=(13,)),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
dl_model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = dl_model.fit(X_train_scaled,y_train,epochs=50,batch_size=16,validation_data=(X_test_scaled, y_test),verbose=1)
```

```python
train_pred_prob = dl_model.predict(X_train_scaled)
train_pred_dl = (train_pred_prob > 0.5).astype(int)


test_pred_prob = dl_model.predict(X_test_scaled)
test_pred_dl = (test_pred_prob > 0.5).astype(int)
```

```python
from sklearn.metrics import accuracy_score

print("DL Train Accuracy :", accuracy_score(y_train, train_pred_dl))
print("DL Test Accuracy  :", accuracy_score(y_test, test_pred_dl))
```

```python
from sklearn.metrics import confusion_matrix

print("Train Confusion Matrix")
print(confusion_matrix(y_train, train_pred_dl))

print("\nTest Confusion Matrix")
print(confusion_matrix(y_test, test_pred_dl))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, test_pred_dl))
```
