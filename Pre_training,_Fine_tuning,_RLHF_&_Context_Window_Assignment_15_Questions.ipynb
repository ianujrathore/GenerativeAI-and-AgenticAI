## Question 1

**Differentiate Pre-training and Fine-tuning with examples.**

**Answer:**

Pre-training is the first stage, where a model is trained on a huge amount of general text data from the internet, books, articles, and so on. At this stage, it doesn't know anything about specific tasks, it just learns language patterns, grammar, facts, and general knowledge. This stage takes a lot of time, data, and compute.

Fine-tuning happens after pre-training, where the already trained model is trained further on a smaller, specific dataset to make it good at a particular task. Since the model already knows language in general, fine-tuning just teaches it to specialize.

Example:

A model like GPT is first pre-trained on massive text from the internet, so it learns general language understanding. Later, if a company wants to build a customer support chatbot, they fine-tune that same model on their own support tickets and conversations, so it learns to answer questions specific to their product.
So basically, pre-training builds general knowledge, and fine-tuning shapes that knowledge for a specific purpose.

## Question 2

**Explain self-supervised learning used in LLM pre-training.**

**Answer:**

Self-supervised learning means the model learns without needing manually labeled data, it creates its own labels from the text itself.
Part of the input is hidden or missing, and the model has to predict it using the rest of the text. Since the actual answer is already present in the original text, no human needs to label anything.

Example:

A sentence like "I love playing cricket" can be turned into a training example by hiding "cricket" and asking the model to predict it using the rest of the sentence. The original word becomes the label automatically.
This is what allows LLMs to be trained on massive amounts of raw internet text without needing manual labeling.

## Question 3

**Why is fine-tuning cheaper than pre-training?**

**Answer:**

Pre-training needs huge amounts of data and compute since the model is learning language from scratch. Fine-tuning starts with an already trained model, so it just needs a smaller dataset and less training time to adjust it for a specific task, making it way cheaper overall.

## Question 4

**Differentiate Instruction Tuning and Fine-tuning.**

**Answer:**

Fine-tuning means training an already pre-trained model further on a specific dataset for one particular task, like sentiment analysis or spam detection.
Instruction tuning is actually a type of fine-tuning, but more specific. Here the model is trained on instruction and response pairs, so it learns to follow different kinds of instructions properly, like "summarize this" or "translate this," instead of just completing text randomly.
Example: fine-tuning on movie reviews makes a model good at sentiment analysis only. Instruction tuning on many instruction-answer pairs makes a model good at following all kinds of prompts in general, which is what makes ChatGPT type models actually useful.

## Question 5

**What is RLHF? Explain all three stages.**

**Answer:**

RLHF stands for Reinforcement Learning from Human Feedback. It's a method used to make model outputs more aligned with what humans actually want, not just grammatically correct but genuinely helpful and safe.

Three stages:

* Supervised fine-tuning: The model is first fine-tuned on good quality human written responses to get a decent starting point.

* Reward model training: Humans rank multiple model outputs from best to worst, and this data trains a separate reward model that learns to score responses like a human would.

* Reinforcement learning: The main model is then further trained using this reward model as feedback, adjusting itself to produce responses that score higher, making it more helpful and aligned overall.

## Question 6

**Explain the role of human feedback in RLHF.**

**Answer:**

Human feedback is basically the whole foundation of RLHF, without it the model wouldn't know what a "good" response actually looks like from a human perspective. Instead of just learning to predict the next likely word, the model learns what humans actually prefer, find helpful, or consider safe.

Humans rank different responses generated by the model from best to worst, and this ranking data is used to train a reward model. That reward model then acts like a judge, guiding the main model during reinforcement learning to produce answers that align more with human expectations, tone, and values, rather than just being technically correct.

## Question 7

**What is a context window? Why is it important?**

**Answer:**

A context window is basically the amount of text, measured in tokens, that a model can look at or remember at one time while generating a response. Anything beyond that limit just gets cut off or ignored by the model.

It's important because it decides how much information the model can actually use to understand a conversation or a document. If the context window is small, the model might forget earlier parts of a long conversation or miss important details from a big document. A larger context window lets the model handle longer inputs, like entire articles or long chats, without losing track of what was said earlier.

## Question 8

**A model has an 8K context window. What does it mean?**

**Answer:**

It means the model can process up to 8,000 tokens at a time, this includes both the input you give it and the output it generates together. If the total conversation or text goes beyond 8,000 tokens, the earlier part gets pushed out or ignored since the model simply can't hold onto it anymore.

So basically, anything within that 8K limit, the model can remember and use as context, but anything before that gets forgotten, kind of like a memory limit for that particular conversation or document.

## Question 9

**Compare 8K, 32K and 128K context windows with use cases.**

**Answer:**

An 8K context window can handle around 6 pages of text at once. It's good enough for short conversations, quick questions and answers, or a simple chatbot reply where you don't need the model to remember too much.

A 32K context window can handle around 24 pages of text. This works better for longer conversations, analyzing a few documents together, or writing and reviewing code across multiple files, since it can hold a lot more information at once compared to 8K.

A 128K context window can handle around 96 or more pages of text. This is useful for really heavy tasks like feeding in an entire book, a huge codebase, or long research papers, since the model can understand everything together without needing to break the content into smaller chunks.

## Question 11

**Write code to fine-tune a sentiment model on a custom dataset (outline acceptable).**

**Answer:**

```python
!pip install transformers datasets

import pandas as pd
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer

data = {
    'text': ["I love this product", "This was a waste of money", "Absolutely amazing experience", "Not good at all"],
    'label': [1, 0, 1, 0]
}
df = pd.DataFrame(data)
dataset = Dataset.from_pandas(df)

model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize_function(example):
    return tokenizer(example['text'], padding="max_length", truncation=True)

tokenized_dataset = dataset.map(tokenize_function, batched=True)

model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)

training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    per_device_train_batch_size=8,
    logging_dir="./logs"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset
)

trainer.train()
```

## Question 12

**Debug: Why might an LLM forget earlier parts of a long conversation?**

**Answer:**

This happens because of the context window limit. Every model can only hold a fixed number of tokens at once, so once a conversation crosses that limit, the oldest messages get pushed out to make room for new ones.

LLMs also don't have real memory, they only see whatever text is included in the current request. Once earlier messages fall outside the window, they're basically invisible to the model.

To fix this, you can use a model with a bigger context window, summarize earlier parts of the conversation and feed that back in, or store important details externally and pull them in only when needed.

**Mini Project: Build a domain-specific chatbot using a pretrained model. Describe dataset, fine-tuning strategy, evaluation metrics, and deployment.**

**Answer:**

```python
# Install libraries
!pip -q install transformers datasets sentencepiece accelerate

import pandas as pd
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSeq2SeqLM,
    DataCollatorForSeq2Seq,
    TrainingArguments,
    Trainer
)
```

```python
# Create sample dataset
data = {
    "question": [
        "What is the last date for fee payment?",
        "When does the new semester start?",
        "How can I apply for a scholarship?",
        "Where can I collect my hall ticket?",
        "How do I check my exam results?",
        "Can I get a fee refund?",
        "How do I apply for a bonafide certificate?",
        "Where is the college library located?",
        "How can I contact the placement cell?",
        "How do I update my personal details?"
    ],
    "answer": [
        "The last date for fee payment is thirtieth July.",
        "The new semester starts on first August.",
        "You can apply for a scholarship through the student portal.",
        "Hall tickets can be collected from the examination office.",
        "Exam results are published on the student portal under Results.",
        "Fee refunds are processed within ten working days upon approval.",
        "You can apply for a bonafide certificate through the admin office.",
        "The library is located on the second floor of the main block.",
        "You can contact the placement cell through email or the campus office.",
        "You can update your details from the profile section of the student portal."
    ]
}
df = pd.DataFrame(data)
dataset = Dataset.from_pandas(df)
dataset
```

```python
# Load tokenizer and model
checkpoint = "google/flan-t5-small"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForSeq2SeqLM.from_pretrained(checkpoint)
```

```python
# Preprocessing function
def preprocess(example):
    model_input = tokenizer(
        example["question"],
        max_length=64,
        truncation=True
    )
    labels = tokenizer(
        text_target=example["answer"],
        max_length=64,
        truncation=True
    )
    model_input["labels"] = labels["input_ids"]
    return model_input

tokenized_dataset = dataset.map(preprocess)
```

```python
# Data collator
data_collator = DataCollatorForSeq2Seq(
    tokenizer=tokenizer,
    model=model
)
```

```python
# Fine-tuning setup
training_args = TrainingArguments(
    output_dir="./college_chatbot",
    per_device_train_batch_size=2,
    num_train_epochs=20,
    learning_rate=2e-4,
    logging_steps=1,
    save_strategy="no",
    report_to="none"
)
```

```python
# Train the model
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator
)
trainer.train()
```

```python
# Save the model
model.save_pretrained("college_chatbot_model")
tokenizer.save_pretrained("college_chatbot_model")
```
